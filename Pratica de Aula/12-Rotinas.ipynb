```python
def contar_inteiros(object: list / tuple) -> int:
    contador = 0
    for x in range(len(object)):
        if type(object[x]) == int:
            contador += 1
    print(f"O número de inteiros na lista é: {contador}")
    return contador
```

```python
obj_1 = [1,2,3, 'a', 'b', 'c', 4.5, 5,4, 32]
resultado = contar_inteiros(object = obj_1)
print(resultado)
```
