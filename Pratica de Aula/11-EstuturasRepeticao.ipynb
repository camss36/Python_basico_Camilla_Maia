```python
import pandas as pd
import numpy as np
```

```python
obj_array = np.random.normal(20, 2.5, 40).reshape(8, 5)
obj_lista = list(obj_array[0])
obj_lista
```

```python
conta = 0
for valor in obj_lista:
    print(f"{conta} - {valor}")
    conta += 1
```

```python
for index, valor in enumerate(obj_lista):
    print(f"{index} - {valor}")
```

```python
conta = 0
while conta < len(obj_lista):
    print(f"{conta} - {obj_lista[conta]}")
    conta += 1
```

```python
for num in range(len(obj_lista)):
    print(f"{num} - {obj_lista[num]}")
```

```python
for valor in enumerate(obj_lista):
    for num in range(len(valor)):
        print(f"{valor[num]}")
```

```python
obj_disc = {
    '0': 10,
    '1': 20,
    '2': 30,
    '3': 40,
    '4': 50
}
obj_disc
```

```python
for index in obj_disc.keys():
    print(f"{index} - {obj_disc[index]}")
```

```python
for valores in obj_disc.items():
    print(f"{valores[0]} - {valores[1]}")
```

```python
for valor in obj_disc.values():
    print(f"{valor}")
```

```python
for linha in obj_array:
    print(f"{linha[:3]}")
```

```python
for linha in obj_array[3]:
    print(f"{linha}")
```

```python
for valor in enumerate(obj_array[3]):
    print(f"{valor}")
```

```python
obj_data = pd.DataFrame(obj_array)
obj_data
```

```python
for valor in obj_data:
    print(f"{obj_data[valor]}")
```

```python
for index in obj_data.index:
    if index > 3:
        print(f"{obj_data.iloc[index]}")
```

```python
for col in obj_data.columns:
    print(f"{obj_data[col]}")
```
