## 03 – Escreva o programa em Python:

```python
lista = list(range(5000))
```

```python
pares = 0
impares = 0

for num in lista:
    if num % 2 == 0:
        pares += 1
    else:
        impares += 1

print(f"Num de pares:", pares)
print(f'Num de impares:', impares)
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
contador_pares = 0
contador_impares = 0
indice = 0

while indice < len(lista):
    numero = lista[indice]

    if numero % 2 == 0:
        contador_pares += 1
    else:
        contador_impares += 1

    indice += 1 #quase crashei o computador por esquecer dessa coisa

print(f"Num de pares:", contador_pares)
print(f"Num de impares:", contador_impares)
```

## 07 – Escreva o programa em Python:

```python
import pandas as pd
```

```python
df_genes = pd.read_csv('genes.csv', index_col="Unnamed: 0")
display(df_genes)
```

```python
col_tamanho = df_genes["Tamanho"]
col_tamanho
```

```python
pares_df = 0
impares_df = 0

for num_df in col_tamanho:
    if num_df % 2 == 0:
        pares_df += 1
    else:
        impares_df += 1

print(f"Num de pares:", pares_df)
print(f'Num de impares:', impares_df)
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
contador_pares_df = 0
contador_impares_df = 0
indice_df = 0

while indice_df < len(col_tamanho):
    numero_df = col_tamanho[indice_df]

    if numero_df % 2 == 0:
        contador_pares_df += 1
    else:
        contador_impares_df += 1

    indice_df += 1

print(f"Num de pares:", contador_pares_df)
print(f"Num de impares:", contador_impares_df)
```
