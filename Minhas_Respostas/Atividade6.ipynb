## 02 – Altere os operadores relacionais para modificar o resultado de cada uma das comparações.

<br>('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]

<br>['a',1,True,'b',0,False] == 1
<br>['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
a_1 = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
b_1 = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
c_1 = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
d_1 = ['a',1,True,'b',0,False] > ['a',True,1]
e_1 = ['a',1,True,'b',0,False] == 1
f_1 = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

print(a_1, b_1, c_1, d_1, e_1, f_1)
```
