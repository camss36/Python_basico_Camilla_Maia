## 2 – No exercício 9 da última aula, foram instanciadas 5 listas. Instancie um objeto composto multidimensional da classe lista com estes cinco vetores. Recupere a matriz.

```python
genes = ['BCRA1', 'TP53', 'HLA-A', 'SOD2', 'APOE']
gene_brca1 = [
    len(genes[0]),
    (genes[0][0:2]),
    genes[0].count("A"),
    "hu"in genes[0],
]
print(gene_brca1)

gene_tp53 = [
    len(genes[1]),
    (genes[1][0:2]),
    genes[1].count("A"),
    "hu"in genes[1],
]
print(gene_tp53)

gene_hla_a = [
    len(genes[2]),
    (genes[2][0:2]),
    genes[2].count("A"),
    "hu"in genes[2],
]
print(gene_hla_a)

gene_sod2 = [
    len(genes[3]),
    (genes[3][0:2]),
    genes[3].count("A"),
    "hu"in genes[3],
]
print(gene_sod2)

gene_apoe = [
    len(genes[4]),
    (genes[4][0:2]),
    genes[4].count("A"),
    "hu"in genes[4],
]

lista_genes = [
    gene_brca1,
    gene_tp53,
    gene_hla_a,
    gene_sod2,
    gene_apoe
]
```

```python
print(lista_genes)
```

## 7 – Converta o objeto instanciado no exercício 10 da aula anterior para uma DataFrame Pandas. Nesta DataFrame, os index devem ser os nomes dos genes e cada coluna deve ter um nome que indique seu conteúdo. Recupere a DataFrame.

```python
dicionario_genes = {
    genes[0]: gene_brca1,
    genes[1]: gene_tp53,
    genes[2]: gene_hla_a,
    genes[3]: gene_sod2,
    genes[4]: gene_apoe
}
print(dicionario_genes)

genes_df = pd.DataFrame(dicionario_genes)
genes_df_tp = genes_df.transpose()
display(genes_df_tp)
```

```python
genes_df_tp.rename(columns={
    0: "Tamanho", 
    1: "Início", 
    2: "Contagem de A", 
    3: "Contém hu?"}, 
    inplace=True
    )
display(genes_df_tp)
```

```python
df_csv = pd.read_csv('../Pratica_de_Aula/data.csv')
display(df_csv)
```

```python
entradas = len(df_csv)
tipos = list(df_csv.dtypes)
elementos = df_csv.dtypes.value_counts().tolist()

colunas = [df_csv.iloc[:,0], df_csv.iloc[:,3], df_csv.iloc[:,8]]
medias = [col.mean() for col in colunas]
desvios = [col.std() for col in colunas]
terceiro_quartil = [col.quantile(0.75) for col in colunas]

dic_csv = {
    'Num_entradas': entradas,
    'Medias': medias,
    'Desv_Pad': desvios,
    'Terc_Quartil': terceiro_quartil,
}

display(dic_csv)
```
