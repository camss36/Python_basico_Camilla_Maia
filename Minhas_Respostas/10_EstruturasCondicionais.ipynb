```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

if p and q and r and s and t == True:
    print("O Resultado é Verdadeiro")
else:
    print("O Resultado é Falso")
```

## 04 – Escreva o programa em Python:

```python
import numpy as np
import pandas as pd

coisas = np.random.normal(10, 3, 5)
coisas
```

```python
contagem = (coisas >= 10).sum()
print(contagem)
```

```python
filtro = coisas[coisas >= 10]
print(filtro)
```

```python
if contagem > 0:
    print(f'Possui', contagem,'números maiores que 10')
    print(f'E os números são:', filtro)
else:
    print('Não possui números maiores que 10')
```
