# Estruturas de repetição

```python
array_1 = np.array([10, 20, 30, 40, 50])
array_1
```

```python
contador = 0
while contador < len(array_1):
    print(array_1[contador])
    contador += 1
```

```python
contador = 0
for i in range(len(array_1)):
    print(array_1[contador])
    contador += 1
```

```python
for i in enumerate(array_1):
    print(i[1])
```

```python
for key, value in enumerate(array_1):
    print(key,value)
```
