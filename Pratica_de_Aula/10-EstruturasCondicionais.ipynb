```python
if obj_array[3,1] >= 200:
    resposta = f"Gene analisado é superior a 200 => {obj_array[3,1]}"
elif obj_array[3,1] < 10:
    resposta = f"Gene analisado é inferior a 10 => {obj_array[3,1]}"
else:
    resposta = f"Gene analisado está entre 10 e 200 => {obj_array[3,1]}"
```
